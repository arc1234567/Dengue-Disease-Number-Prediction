```python
#Creating Pearson Correlation matrix 
import matplotlib.pyplot as plt #Importing matplotlib library
import seaborn as sns#importing seaborn
import statsmodels.api as sm
plt.figure(figsize=(12,10))
cor = df.corr()
sns.heatmap(cor, annot=True,cmap=plt.cm.Reds)
plt.show()
```

```python
# Scaling on all required columns
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['precipitation_amt_mm'] = scaler.fit_transform(df['precipitation_amt_mm'].values.reshape(-1,1))
df['reanalysis_air_temp_k'] = scaler.fit_transform(df['reanalysis_air_temp_k'].values.reshape(-1,1))
df['reanalysis_avg_temp_k'] = scaler.fit_transform(df['reanalysis_avg_temp_k'].values.reshape(-1,1))
df['reanalysis_dew_point_temp_k'] = scaler.fit_transform(df['reanalysis_dew_point_temp_k'].values.reshape(-1,1))
df['reanalysis_max_air_temp_k'] = scaler.fit_transform(df['reanalysis_max_air_temp_k'].values.reshape(-1,1))

df['reanalysis_min_air_temp_k'] = scaler.fit_transform(df['reanalysis_min_air_temp_k'].values.reshape(-1,1))
df['reanalysis_precip_amt_kg_per_m2'] = scaler.fit_transform(df['reanalysis_precip_amt_kg_per_m2'].values.reshape(-1,1))
df['reanalysis_relative_humidity_percent'] = scaler.fit_transform(df['reanalysis_relative_humidity_percent'].values.reshape(-1,1))
df['reanalysis_sat_precip_amt_mm'] = scaler.fit_transform(df['reanalysis_sat_precip_amt_mm'].values.reshape(-1,1))
df['reanalysis_specific_humidity_g_per_kg'] = scaler.fit_transform(df['reanalysis_specific_humidity_g_per_kg'].values.reshape(-1,1))

df['reanalysis_tdtr_k'] = scaler.fit_transform(df['reanalysis_tdtr_k'].values.reshape(-1,1))
df['station_avg_temp_c'] = scaler.fit_transform(df['station_avg_temp_c'].values.reshape(-1,1))
df['station_diur_temp_rng_c'] = scaler.fit_transform(df['station_diur_temp_rng_c'].values.reshape(-1,1))
df['station_max_temp_c'] = scaler.fit_transform(df['station_max_temp_c'].values.reshape(-1,1))
df['station_min_temp_c'] = scaler.fit_transform(df['station_min_temp_c'].values.reshape(-1,1))
df['station_precip_mm'] = scaler.fit_transform(df['station_precip_mm'].values.reshape(-1,1))
```
